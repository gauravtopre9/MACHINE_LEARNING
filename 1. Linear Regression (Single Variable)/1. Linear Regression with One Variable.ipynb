# Linear Regression with One Variable  (Univariate)

**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

**Importing Data**

- First Column Contains area of House 
- Second Column Contains Price of the house

```python
df = pd.read_csv('Book1.csv')
df.head()
```

**Plotting Initial Data ( Scatter Plot )**

```python
plt.scatter(df.area,df.Price, color = 'red', marker='+')

plt.title('Scatter Plot', fontsize = 14)
plt.xlabel('Area of House')
plt.ylabel('Price of House')
```

**Creating a LinearRegression Object**

```python
reg = linear_model.LinearRegression()
```

**Fitting Data into a Reg Object**

- We are going to fit area as 2D array
- and price from df

```python
reg.fit(df[['area']],df.Price)
```

**Now our model is Ready to Predict the Prices of House**

- We are trying to predict Price for house with area 3800

```python
reg.predict([[3800]])
```

**The above Equation had helped us in making this prediction which we have made. Reg model had calculated 'm' which is Coeff and the 'b' which is intercept internally let's see what is the value of intercept and coeffecient**

```python
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
y = m*3800 + b #fitting coeff and intercept in line equation
```

```python
y
```

**If we try this y for all the values of area in our Dataframe we will be getting the following Scatter plot whre the Blue line is nothing but the Regression Line.**

- For this Regression Line the error between the line and the datapoint is minimum

```python
reg.predict([[5000]])
```

**Now Let's Predict the prices for random values of Areas**

```python
d = pd.read_csv('sample predict.csv')
d
```

```python
p = reg.predict(d[['area']])
```

```python
d['Predicted_Price'] = p
```

```python
d
```

```python
d.to_csv('predictions.csv', index=False)
```

```python
plt.scatter(d.area,d.Predicted_Price, color = 'red', marker='+')

plt.title('Scatter Plot with Regression Line', fontsize = 14)
plt.xlabel('Area of House')
plt.ylabel('Price of House')
plt.plot(d.area,reg.predict(d[['area']]))
```

# Exercise on per capita income(canada)

```python
df1 = pd.read_csv('per capita income.csv')
df1.head()
```

```python
df1.drop('Unnamed: 2', axis=1, inplace = True)
```

```python
df1.head()
```

```python
plt.scatter(df1.year,df1.pci, color = 'red', marker = '+')
plt.title('Scatter Plot without Regression Line', fontsize = 16)
plt.xlabel('Year')
plt.ylabel('Per Capita Income (PCI)')
```

```python
reg_1 = linear_model.LinearRegression()
```

```python
reg_1.fit(df1[['year']],df1.pci)
```

```python
m = reg_1.coef_
b = reg_1.intercept_

m,b
```

```python
reg_1.predict([['2022']])
```

```python
plt.scatter(df1.year,df1.pci, color = 'red', marker = '+')
plt.plot(df1.year,reg_1.predict(df1[['year']]))
         
plt.title('Scatter Plot with Regression Line', fontsize = 16)
plt.xlabel('Year')
plt.ylabel('Per Capita Income (PCI)')
```
