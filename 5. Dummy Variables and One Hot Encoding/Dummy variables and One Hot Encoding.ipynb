# Categorical Variables 

- Nominal Variables : Categorical variables which does not have any numerical significance 

e.g: male,female



- Ordinal Variables : Categorical Variables that have some numberical Significance 

e.g: Good,better, best

```python
dummies = pd.get_dummies(df.town)
```

```python
dummies
```

```python
merged = pd.concat([df,dummies], axis = 1)
merged
```

```python
merged.drop('town', axis = 1, inplace = True)
```

```python
merged
```

**We are going to drop one of the dummy variable column on the basis of solution to dummy variable trap**

```python
merged.drop('west windsor', axis = 1, inplace = True)
```

```python
merged
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
x = merged.drop('price', axis =1)
x
```

```python
y = merged.price
y
```

```python
reg.fit(x,y)
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.predict([[3400,0,0]])
```

```python
reg.score(x,y)
```

# One Hot Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le =  LabelEncoder()
```

```python
dfle = df

le.fit_transform(df.town)
```

```python
df.town = le.fit_transform(df.town)
```

```python
df
```

```python
x = df[['town', 'area']].values
x
```

```python
y = df.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town',OneHotEncoder(),[0])], remainder = 'passthrough')
```

```python
x1 = ct.fit_transform(x)
```

```python
x1
```

```python
x1 = x1[:,1:]
x1
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x1,y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[1,0,3400]])
```

```python
model.score(x1,y)
```
